```python
np.base_repr(1,base =  4)[0]

def return_b(x,bit):
    bit_rep = np.base_repr(x, bit)
    ans = "0"*(4-len(bit_rep)) + bit_rep
    return ans

print(return_b(8, 4)[3])
```

```python
P = np.zeros((256,256))
for i in range(P.shape[0]):
    for j in range(P.shape[1]):
        if(return_b(i,4)[2] == return_b(j,4)[3] and return_b(i,4)[1] == return_b(j,4)[2] and return_b(i,4)[0] == return_b(j,4)[1]):
            P[i][j] = 0.25
            
P[228][:] = 0
P[228][228]= 1
print(np.sum(P[228][:]))
```

```python
R = -1*np.ones((256,1))
R[228][0]= 0
```

```python
I = np.identity(256)
V = np.dot(np.linalg.pinv(I-P),R)
```

```python
print(V.shape)
print(np.amin(V))
print(np.amax(V))

print(np.where(V == -1))
print(return_b(0,4))
```
