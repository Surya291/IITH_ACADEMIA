```python
#### Q1. NMOS : 


f1 = '1.a.n.1.dat'

x = np.loadtxt(f1,usecols = 1)
y = np.loadtxt(f1,usecols = 3) * 1e-3

v = [0.9, 1.8]
r = [4.63, 8.75]
plt.scatter(v,r)


plt.plot(x,y)

plt.xlabel('Vds (in volts)')
plt.ylabel('R : (Vds/Id) : in kOhms ')
plt.title(" R of NMOS wrt Vds: for  W/L = 240/180")

plt.show()
plt.figure()
```

```python

f1 = '5.a.2.dat'
x = np.loadtxt(f1,usecols = 0)
y = np.loadtxt(f1,usecols = 1)   

plt.scatter(0.64, -9.43,color = 'r')
plt.annotate("Peakgain = -9.43", (0.8, -9.43))

plt.plot(x,y)
plt.title("Gain vs Vin")
plt.xlabel("Vin (in V)")
plt.ylabel("Gain (dVout / dVin) ")
plt.figure()
```

```python
f1 = '5.b.1.dat'
x = np.loadtxt(f1,usecols = 0)
y = np.loadtxt(f1,usecols = 1)   
x_t = np.split(x,3)
y_t = np.split(y,3)



vm = [ 7.228527e-01, 6.278440e-01, 5.921920e-01]



plt.title('Vout vs Vin \n for different Rl')
plt.xlabel("Vin (in V)")
plt.ylabel("Gain (dVout / dVin) ")
plt.scatter(vm,vm)
plt.plot(x_t[0],y_t[0])
plt.plot(x_t[1],y_t[1])
plt.plot(x_t[2],y_t[2])
plt.legend(['Rl = 50k Ohm','Rl = 120k Ohms', 'Rl = 190k Ohm'])
plt.grid()
```
