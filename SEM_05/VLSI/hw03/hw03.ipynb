```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#### Q1 NOLOAD


f1 = '1.a.1.dat'

x = np.loadtxt(f1,usecols = 0)*1e9
y = np.loadtxt(f1,usecols = 1) 


plt.plot(y,x)

plt.xlabel('S (scaling factor)')
plt.ylabel(' Propagation delay (ns) ')
plt.title("Prop delay in Load : C = 20p")

plt.show()
plt.figure()
```

```python
#### Q1 NOLOAD

f1 = '1.a.2.dat'

x = np.loadtxt(f1,usecols = 0)* 1e+12
y = np.loadtxt(f1,usecols = 1) 


plt.plot(y,x)

plt.xlabel('S (scaling factor)')
plt.ylabel(' Propagation delay (ps) ')
plt.title("Prop delay in  NOLOAD")

plt.show()
plt.figure()
```

```python
#### Q1 LOAD


f1 = '1.b._wp.dat'

x = np.loadtxt(f1,usecols = 0)*1e9
thl = np.loadtxt(f1,usecols = 1) *1e9
tlh = np.loadtxt(f1,usecols = 3) *1e9
tp = np.loadtxt(f1,usecols = 5) *1e9



plt.plot(x,thl)
plt.plot(x,tlh)
plt.plot(x,tp)

plt.legend(['t_hl','t_lh', 't_p' ])

plt.xlabel('Wp (in nm)')
plt.ylabel(' delay (ns) ')
plt.title("Prop delay for Load : C = 20p while changing Wp (keeping Wn/Ln = 1)")

plt.show()
plt.figure()
```
