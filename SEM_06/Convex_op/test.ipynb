```python
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt
```

```python
f = np.array([[1,0,2,1]]).T
x = np.array([[0,1,2,3]]).T
```

```python

f_cap = []
for i in range(4):
    
    m = cp.Variable((1,1))
    c = cp.Variable((1,1))

    obj = cp.Minimize(m*x[i][0] + c)

    constraint = [f<= m*x + c]

    prob = cp.Problem(obj, constraint)
    prob.solve()
    
    f_cap.append(obj.value)
plt.plot(x,f)
plt.plot(x,f_cap)
```

```python

f_cap = []
for i in range(4):
    
    alpha = cp.Variable((4,1))

    obj = cp.Minimize(alpha.T@f)

    constraint = [alpha.T@x == x[i][0], sum(alpha) == 1, alpha >=0]

    prob = cp.Problem(obj, constraint)
    prob.solve()
    
    f_cap.append(obj.value)
plt.plot(x,f)
plt.plot(x,f_cap)
```
