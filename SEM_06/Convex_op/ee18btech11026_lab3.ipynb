###  1.a No constraint on w
### This leads to a state where some elements of w become negative , which means that we inturn deposit money rather than withdrawing resulting in less initial investment

```python
import numpy as np
import cvxpy as cp

### Declaring variable : 
b = cp.Variable((n,1))
w = cp.Variable((n,1))
x = cp.Variable((m,1))
i = np.ones((m,1))

### Objective ..
obj = cp.Minimize(b[0] + (i.T)@x )

b_fut = b[1:]
b_pre = b[0:n-1]
w_pre = w[0:n-1]

print(b_pre.shape)
### Constraints : 
constraints = [b >= 0 , w+(P@x) >= e , x>=0 , b_fut == ((1+rho)*b_pre - w_pre)]
prob = cp.Problem(obj, constraints)
prob.solve()

print("------------------------  Minimum investment required  with NO constraint on w----------------------\n")
print((b[0] + (i.T)@x).value)
```

###  1.b Imposing a constraint on w>=0
###  This means we intend to only withdraw money, thus the inital investment wil be greater than previous

```python
import numpy as np
import cvxpy as cp

### Declaring variable : 
b = cp.Variable((n,1))
w = cp.Variable((n,1))
x = cp.Variable((m,1))
i = np.ones((m,1))

### Objective ..
obj = cp.Minimize(b[0] + (i.T)@x )

b_fut = b[1:]
b_pre = b[0:n-1]
w_pre = w[0:n-1]

print(b_pre.shape)
### Constraints : 
constraints = [b >= 0 , w+(P@x) >= e , x>=0 , b_fut == ((1+rho)*b_pre - w_pre),  w>=0]
prob = cp.Problem(obj, constraints)
prob.solve()

print("------------------------  Minimum investment required with w>=0 constraint ----------------------\n")
print((b[0] + (i.T)@x).value)
```

### 2. No investment  :  x = 0 , constraints on w doesnot matter , Since w will be >=0 to satisfy the objective
